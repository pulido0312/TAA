```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import time
```

```python
X,y=fetch_openml(data_id='1560', return_X_y=True)
```

```python
X = MinMaxScaler().fit_transform(X)
X.shape, y.shape
```

<div class="alert alert-block alert-success">

<b>skf = StratifiedKFold(n_splits=10):</b> Crea un objeto StratifiedKFold que se utilizará para realizar la validación cruzada estratificada con 10 folds. La validación cruzada estratificada garantiza que las proporciones de clases se mantengan en cada fold.

<b>results_summary = np.zeros((10,3), dtype='float'):</b> Crea una matriz de ceros de tamaño (10, 3) que se utilizará para almacenar los resultados de precisión y tiempos de ejecución para cada tamaño de capa oculta y fold.

<b>size = 0:</b> Inicializa la variable size en 0, que se utilizará para indexar la matriz results_summary.

<b>El bucle for i in range(5,51,5):</b> itera sobre diferentes tamaños de capa oculta, desde 5 hasta 50 en incrementos de 5.

<b>results_per_size = np.zeros((10,3), dtype='float'):</b> Crea una matriz de ceros de tamaño (10, 3) para almacenar los resultados de precisión y tiempos de ejecución para el tamaño de capa oculta actual.

<b>print("\033[1mTamaño de capa:\033[0m",i):</b>  Imprime el tamaño de capa oculta actual.

<b>El bucle for train_index, test_index in skf.split(X,y):</b> realiza la validación cruzada estratificada dividiendo los datos en conjuntos de entrenamiento y prueba.

<b>X_train, X_test = X[train_index], X[test_index]:</b> Divide los datos de características en conjuntos de entrenamiento y prueba según los índices generados por skf.split.

<b>y_train, y_test = y[train_index], y[test_index]:</b> Divide las etiquetas de destino en conjuntos de entrenamiento y prueba según los índices generados por skf.split.

<b>start = time.time():</b> Registra el tiempo de inicio del entrenamiento del clasificador.

<b>clf = MLPClassifier(hidden_layer_sizes=(i,), max_iter=1000):</b> Crea un objeto MLPClassifier con el tamaño de capa oculta actual (especificado por i) y un número máximo de iteraciones de 1000.

<b>clf.fit(X_train, y_train):</b> Entrena el clasificador MLP utilizando los conjuntos de entrenamiento.

<b>predict_train = clf.predict(X_train):</b> Realiza predicciones en los datos de entrenamiento.

<b>predict_test = clf.predict(X_test):</b> Realiza predicciones en los datos de prueba.

<b>accuracy_score_train = accuracy_score(y_train, predict_train):</b> Calcula la precisión en los datos de entrenamiento comparando las etiquetas verdaderas (y_train) con las predicciones (predict_train).

<b>accuracy_score_test = accuracy_score(y_test, predict_test):</b> Calcula la precisión en los datos de prueba comparando las etiquetas verdaderas (y_test) con las predicciones (predict_test).

<b>end = time.time():</b> Registra el tiempo de finalización del bucle

```python
skf = StratifiedKFold(n_splits=10)
results_summary = np.zeros((10,3), dtype='float')
size = 0
for i in range(5,51,5):
    results_per_size = np.zeros((10,3), dtype='float')
    k = 0
    print("\033[1mTamaño de capa:\033[0m",i)
    
    for train_index, test_index in skf.split(X,y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        start = time.time()
        
        clf = MLPClassifier(hidden_layer_sizes=(i,), max_iter=1000)
        clf.fit(X_train, y_train)
        
        predict_train = clf.predict(X_train)
        predict_test = clf.predict(X_test)
        
        accuracy_score_train = accuracy_score(y_train, predict_train)
        accuracy_score_test = accuracy_score(y_test, predict_test)
        
        end = time.time()
        execution_time = end - start
        
        results_per_size[k] = [accuracy_score_train,accuracy_score_test,execution_time]
        k += 1
    
    print('\33[4mFold     Entrena     Verifica     T.Ejecución\33[0m')
    for i in range(10):
        print(i,'\t %.4f'%results_per_size[i][0],'     %.4f'%results_per_size[i][1],'      %.4f'%results_per_size[i][2])   
    print('\n')
    
    mean_train = np.mean(results_per_size[:,0], axis = 0, dtype=np.float64)
    mean_test = np.mean(results_per_size[:,1], axis = 0, dtype=np.float64)
    mean_execution_time = np.sum(results_per_size[:,2], axis = 0, dtype=np.float64)
    
    print('MEDIA (Entrena)= %.4f' %mean_train)
    print('MEDIA (Verifica)= %.4f' %mean_test)
    print('Tiempo Total (seg)= %.4f' %mean_execution_time)
    print('\n')
    
    results_summary[size] = [mean_train, mean_test, mean_execution_time]
    size += 1
```

```python
# Impresión de los resultados para cada tamaño de capa oculta
size = 5
print('\33[4mNo.Ocultas     Entrena     Verifica     T.Ejecución\33[0m')
for i in range(10):
    print(size,'\t       %.4f'%results_summary[i][0],'     %.4f'%results_summary[i][1],'      %.4f'%results_summary[i][2])
    size +=5
```

```python
# Mayor precisión media en el test
best_ac = np.argmax(results_summary[:,1], axis=0)
print("Entendiendo mejor resultado como la mayor precisión media en la prueba del clasificador, obtenemos este con un tamaño de capa oculta de:",(best_ac+1)*5,'(Verifica: %.4f)'%results_summary[best_ac][1])
```

```python
# Menor tiempo de ejecución
best_t = np.argmin(results_summary[:,2], axis=0)
print("El resultado que menor tiempo conlleva de ejecución lo obtenemos con un tamaño de capa de:",(best_t+1)*5, '(T.Ejecución: %.4f)'%results_summary[best_t][2])
```
